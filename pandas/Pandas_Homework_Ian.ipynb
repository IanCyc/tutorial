```python
import datetime
```

```python
df['date'] = df['context_timestamp'].apply(lambda x: datetime.datetime.fromtimestamp(int(x)).date())
```

```python
df2 = df[['item_id', 'date','is_trade']]
```

```python
df_sts = pd.DataFrame({})
for _end in df2['date'].unique():
    _begin = _end - datetime.timedelta(3)
    df3 = df2[(_begin <= df2['date']) & (df2['date'] < _end)]
    df3_buy = df3.groupby('item_id').sum()[['is_trade']]
    df3_buy = df3_buy.rename(columns={'is_trade':'buy'})
    df3_click = df3.groupby('item_id').count()[['is_trade']]
    df3_click = df3_click.rename(columns={'is_trade':'3days_click'})
    
    df3_buy['date'] = pd.Series([_end for i in range(df3_buy.shape[0])], index = df3_buy.index)
    #df3_buy['his_rat'] = df3_buy['buy']/df3_click['3days_click']
    df3_buy['click'] = df3_click['3days_click']
    df3_buy['his_rate'] = df3_buy['buy']/df3_buy['click']
    #df3_buy['rate'] = df3_buy.apply(lambda x: x.buy/x.click,axis=1)
    df_sts = pd.concat([df_sts, df3_buy], axis=0)
df_sts['item_id'] = df_sts.index
df_sts.shape
    
```

```python
df_final_sts = df_sts.drop(['buy','click'], axis=1)
```

```python
df_final_sts.head()
```

```python
df = df.merge(df_final_sts, on=['item_id','date'], how='left')
df.head()
```
